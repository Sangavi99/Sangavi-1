```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input,Dense
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data.drop(['RowNumber','CustomerId','Surname'], axis=1,inplace=True)
data
```

```python
col=['Geography','Gender']
labelencoder = LabelEncoder()
for i in col:
    data[i] = labelencoder.fit_transform(data[i])
```

```python
data['Geography'].unique()
```

```python
data['Geography'] = labelencoder.fit_transform(data['Geography'])
data
```

```python
x=data.iloc[:,:-1]#data[data.columns[:-1]]
y=data.iloc[:,-1]
```

```python
scaler = StandardScaler()#-3 to 3
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_data=pd.DataFrame(x_scaled,columns=data.columns[:-1])
x_data
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y, test_size=0.25, random_state=0)
```

```python
model = Sequential()
model.add(Input(shape=(30,)))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
from tensorflow.keras.optimizers import SGD
model.compile(optimizer=SGD(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train,validation_data=(X_test,y_test),batch_size=32, epochs=25)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
